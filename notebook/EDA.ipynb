# Data Cleaning 

```python
cleaned_df1 = df.copy()  
```

```python
cleaned_df1["Invoice"] = cleaned_df1["Invoice"].astype("str")

mask = (cleaned_df1["Invoice"].str.match("^\\d{6}$")==True)  # mask is use as filtering expression to filter unwanted invoice 

cleaned_df1 = cleaned_df1[mask]

cleaned_df1
```

```python
cleaned_df1["StockCode"] = cleaned_df1["StockCode"].astype("str")

mask = (
    (cleaned_df1["StockCode"].str.match("^\\d{5}$") == True) # Condition 1
    | (cleaned_df1["StockCode"].str.match("^\\d{5}[a-zA-Z]+$") == True) # Condition 2
    | (cleaned_df1["StockCode"].str.match("^PADS$") == True) # Condition 3
)

cleaned_df1 = cleaned_df1[mask]

cleaned_df1
```

```python
cleaned_df1 = cleaned_df1.drop_duplicates()
```

```python
cleaned_df1.dropna(subset=["Customer ID"], inplace=True)
```

```python
cleaned_df1.describe()
```

```python
cleaned_df1["Price"] = pd.to_numeric(cleaned_df1["Price"],errors="coerce") # convert the Price to numeric as it show the data type is string
```

```python
cleaned_df1 = cleaned_df1[cleaned_df1["Price"] > 0] # To filter all unwanted 0
```

```python
cleaned_df1 = cleaned_df1[cleaned_df1["Quantity"]>0] # TO filter all unwanted quantity with 0 or less than 0 
```
